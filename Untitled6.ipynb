```python
import pandas as pd

# Sample synthetic dataset
data = {
    'EmployeeID': range(1, 11),
    'Age': [25, 45, 37, 50, 29, 41, 35, 28, 52, 47],
    'BMI': [22, 31, 28, 35, 24, 30, 27, 25, 36, 33],
    'HeartRate': [72, 88, 75, 95, 70, 90, 80, 74, 100, 85],
    'WorkingHoursPerWeek': [40, 55, 50, 60, 38, 45, 48, 42, 62, 57],
    'SleepHours': [7, 5, 6, 4, 8, 5, 6, 7, 4, 5],
    'StepsPerDay': [7000, 3000, 5000, 2000, 8000, 4000, 6000, 7500, 1500, 3500]
}

df = pd.DataFrame(data)
df.head()
```

```python
from scipy.stats import zscore
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Detect anomalies using Z-score
z_scores = df[['BMI', 'HeartRate', 'SleepHours', 'StepsPerDay']].apply(zscore)
df['Anomaly'] = (z_scores.abs() > 2).any(axis=1)

# Clustering using KMeans
features = ['BMI', 'HeartRate', 'WorkingHoursPerWeek', 'SleepHours', 'StepsPerDay']
X = df[features]
X_scaled = StandardScaler().fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=0)
df['RiskGroup'] = kmeans.fit_predict(X_scaled)

# Label risk levels based on averages
cluster_centers = kmeans.cluster_centers_.mean(axis=1)
labels = pd.Series(cluster_centers).sort_values().index
risk_map = {labels[0]: 'Low', labels[1]: 'Medium', labels[2]: 'High'}
df['RiskGroup'] = df['RiskGroup'].map(risk_map)
```

```python
def suggest(row):
    if row['RiskGroup'] == 'High':
        return 'Health checkup, reduce work hours, improve sleep'
    elif row['RiskGroup'] == 'Medium':
        return 'Encourage fitness, monitor heart rate'
    else:
        return 'Maintain routine, encourage healthy habits'

df['SuggestedIntervention'] = df.apply(suggest, axis=1)
df[['EmployeeID', 'RiskGroup', 'SuggestedIntervention']]
```

```python
import matplotlib.pyplot as plt

# Summary
print(df['RiskGroup'].value_counts())

# Bar chart
df['RiskGroup'].value_counts().plot(kind='bar', title='Risk Group Distribution')
plt.xlabel('Risk Group')
plt.ylabel('Employees')
plt.show()
```
